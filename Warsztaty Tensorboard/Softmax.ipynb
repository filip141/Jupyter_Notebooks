```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

After importing tensorflow import MNIST dataset

```python
mnist = input_data.read_data_sets('MNIST_data', one_hot=True)
```

Define input

```python
input_shape = 784
output_shape = 10
x = tf.placeholder(tf.float32, [None, input_shape], name="network_input")
```

Create softmax classifier weights and biases

```python
W = tf.Variable(tf.truncated_normal([input_shape, output_shape]))
b = tf.Variable(tf.truncated_normal([output_shape]))
```

Define network and its output

```python
logits = tf.matmul(x, W) + b
net_out = tf.nn.softmax(logits)
```

Define output placeholder for labels

```python
y = tf.placeholder(tf.float32, [None, output_shape], name="labels")
```

Define cross entropy

```python
cross_ent = -tf.reduce_sum(y * tf.log(net_out))
```

Define accuracy

```python
is_correct = tf.equal(tf.argmax(net_out, 1), tf.argmax(y, 1))
acc = tf.reduce_mean(tf.cast(is_correct, tf.float32))
```

Define Optimizer

```python
opt = tf.train.GradientDescentOptimizer(0.001)
loss = opt.minimize(cross_ent)
```

Define session and init variables

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

```python
# Run main loop
for epoch_idx in range(0, 200):
    for sample_idx in range(0, 500):
        batch_x, batch_y = mnist.train.next_batch(100)
        sess.run(loss, feed_dict={y: batch_y, x: batch_x})
        
        train_acc, train_loss = sess.run([acc, cross_ent], feed_dict={y: batch_y, x: batch_x})
        print("Train Accuracy: {}, Train Loss: {}".format(train_acc, train_loss))
        test_acc, test_loss = sess.run([acc, cross_ent], feed_dict={y: mnist.test.labels, x: mnist.test.images})
        print("Test Accuracy: {}, Test Loss: {}".format(test_acc, test_loss))
```
