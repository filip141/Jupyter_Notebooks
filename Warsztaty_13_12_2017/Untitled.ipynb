```python
import tensorflow as tf
from tensorflow.examples.tutorials.mnist import input_data
```

Download MNIST Dataset

```python
mnist = input_data.read_data_sets("MNIST_data", one_hot=True)
```

Define Layer sizes

```python
input_shape = 784
output_shape = 10
hidden_1 = 300
hidden_2 = 200
hidden_3 = 60
hidden_4 = 30
```

Define input placeholder

```python
x = tf.placeholder(tf.float32, [None, input_shape], name="network_input")
```

Define variables for layers

```python
W_1 = tf.Variable(tf.truncated_normal([input_shape, hidden_1], stddev=0.1))
b_1 = tf.Variable(tf.zeros(hidden_1))

W_2 = tf.Variable(tf.truncated_normal([hidden_1, hidden_2], stddev=0.1))
b_2 = tf.Variable(tf.zeros(hidden_2))

W_3 = tf.Variable(tf.truncated_normal([hidden_2, hidden_3], stddev=0.1))
b_3 = tf.Variable(tf.zeros(hidden_3))

W_4 = tf.Variable(tf.truncated_normal([hidden_3, hidden_4], stddev=0.1))
b_4 = tf.Variable(tf.zeros(hidden_4))

W_5 = tf.Variable(tf.truncated_normal([hidden_4, output_shape], stddev=0.1))
b_5 = tf.Variable(tf.zeros(output_shape))
```

Most interesting think, define network

```python
act_1 = tf.nn.relu(tf.matmul(x, W_1) + b_1)
act_2 = tf.nn.relu(tf.matmul(act_1, W_2) + b_2)
act_3 = tf.nn.relu(tf.matmul(act_2, W_3) + b_3)
act_4 = tf.nn.relu(tf.matmul(act_3, W_4) + b_4)
net_logits = tf.matmul(act_4, W_5) + b_5
```

Define labels

```python
y = tf.placeholder(tf.float32, [None, output_shape], name="labels")
```

Define Crossentropy

```python
cross_ent = tf.reduce_mean(tf.nn.softmax_cross_entropy_with_logits(logits=net_logits, labels=y))
```

**Accuracy**

```python
is_correct = tf.equal(tf.argmax(net_logits, 1), tf.argmax(y, 1))
acc = tf.reduce_mean(tf.cast(is_correct, tf.float32))
```

## Optimizer !! ;) ;) ;)

```python
opt = tf.train.AdamOptimizer(0.003)
loss = opt.minimize(cross_ent)
```

```python
sess = tf.Session()
init = tf.global_variables_initializer()
sess.run(init)
```

# Main program Loop

```python
for epoch_idx in range(0, 300):
    for sample_idx in range(0, 500):
        batch_x, batch_y = mnist.train.next_batch(100)
        
        sess.run(loss, feed_dict={x: batch_x, y: batch_y})
        
        train_acc, train_loss = sess.run([acc, cross_ent], feed_dict={x: batch_x, y: batch_y})
        print("Train accuracy: {}, Loss: {}".format(train_acc, train_loss))
        
        
        test_acc, test_loss = sess.run([acc, cross_ent], feed_dict={x: mnist.test.images, y: mnist.test.labels})
        print("Test accuracy: {}, Loss: {}".format(test_acc, test_loss))
```
